# 【問題3】Iris（2値分類）をKerasで学習

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
y = data.target[50:] - 1 
X = data.data[50:]
```

```python
# trainとtestに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# さらにtrainとvalに分割
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)

K.clear_session()
model = keras.Sequential()
model.add(layers.Dense(50, activation='relu', input_shape=(X_train.shape[1], )))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
print(model.summary())
history = model.fit(X_train, y_train, batch_size=10, epochs=20, verbose=1, validation_data=(X_train, y_train))
y_pred = model.predict(X_val)
y_pred_label = np.where(y_pred>=0.5, 1, 0)
print(y_pred_label.reshape(-1))
print(y_val)
```

# 【問題4】Iris（多値分類）をKerasで学習

```python
y = data.target
X = data.data
y = to_categorical(y)
# trainとtestに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# さらにtrainとvalに分割
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
```

```python
K.clear_session()
model = keras.Sequential()
model.add(layers.Dense(50, activation='relu', input_shape=(X_train.shape[1], )))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
print(model.summary())
history = model.fit(X_train, y_train, batch_size=10, epochs=20, verbose=1, validation_data=(X_train, y_train))
y_pred = model.predict_classes(X_val)
print(y_pred)
print(np.argmax(y_val,axis=1))
```

# 【問題5】House PricesをKerasで学習

```python
df = pd.read_csv('train.csv')
```

```python
y = np.array(df['SalePrice'])
X = np.array(df[['GrLivArea', 'YearBuilt']])
y = np.log(y)
# trainとtestに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# さらにtrainとvalに分割
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
scaler= StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

```python
K.clear_session()
model = keras.Sequential()
model.add(layers.Dense(50, activation='relu', input_shape=(X_train.shape[1], )))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(1))
model.compile(loss='mse', optimizer='Adam', metrics=['mse'])
print(model.summary())
history = model.fit(X_train, y_train, batch_size=10, epochs=20, verbose=1, validation_data=(X_train, y_train))
y_pred = model.predict(X_val)
print(y_pred[:10].reshape(-1))
print(y_val[:10])
```

# 【問題6】MNISTをKerasで学習

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train.astype(np.float)
X_test = X_test.astype(np.float)
X_train /= 255
X_test /= 255
print(X_train.max()) # 1.0
print(X_train.min()) # 0.0
```

```python
X_train = X_train[:, :, :, np.newaxis]
X_test = X_test[:, :, :, np.newaxis]
print(X_train.shape)
print(X_test.shape)
```

```python
y_train_one_hot = to_categorical(y_train[:, np.newaxis])
y_test_one_hot = to_categorical(y_test[:, np.newaxis])
print(y_train.shape) # (60000,)
print(y_train_one_hot.shape) # (60000, 10)
print(y_train_one_hot.dtype) # float64
```

```python
# trainとvalに分割
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train_one_hot, test_size=0.2, random_state=0)
```

```python
K.clear_session()
model = keras.Sequential()
model.add(layers.Conv2D(4, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(8, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(200, activation='softmax'))
model.add(layers.Dense(10, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
print(model.summary())
history = model.fit(X_train, y_train, batch_size=20, epochs=5, verbose=1, validation_data=(X_train, y_train))
y_pred = model.predict_classes(X_val)
print(y_pred)
print(np.argmax(y_val,axis=1))
```
