# 【問題6】図の作成

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

```python
df['Species'].value_counts()
```

```python
df.replace({'Species': {0: data['target_names'][0], 1: data['target_names'][1], 2: data['target_names'][2] }}, inplace=True)
```

```python
labels = [df['Species'].unique()[i] for i in range(df['Species'].nunique())]
print(labels)
```

```python
sizes = [df['Species'].value_counts()[label] for label in labels]
print(sizes)
```

```python
features = data['feature_names']
print(features)
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
plt.show()
```

## 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する

```python
sepal_length = df[features[0]]
sepal_width = df[features[1]]
petal_length = df[features[2]]
petal_width = df[features[3]]
```

```python
fig,ax = plt.subplots(1,4, figsize=(12, 4))
ax[0].boxplot([sepal_length[df['Species']==labels[0]], 
               sepal_length[df['Species']==labels[1]],
               sepal_length[df['Species']==labels[2]]], labels=[labels[0], labels[1], labels[2]])
ax[0].set_title(features[0])
ax[1].boxplot([sepal_width[df['Species']==labels[0]], 
               sepal_width[df['Species']==labels[1]],
               sepal_width[df['Species']==labels[2]]], labels=[labels[0], labels[1], labels[2]])
ax[1].set_title(features[1])
ax[2].boxplot([petal_length[df['Species']==labels[0]], 
               petal_length[df['Species']==labels[1]],
               petal_length[df['Species']==labels[2]]], labels=[labels[0], labels[1], labels[2]])
ax[2].set_title(features[2])
ax[3].boxplot([petal_width[df['Species']==labels[0]], 
               petal_width[df['Species']==labels[1]],
               petal_width[df['Species']==labels[2]]], labels=[labels[0], labels[1], labels[2]])
ax[3].set_title(features[3])
plt.show()
```

## 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

```python
fig,ax = plt.subplots(1,4, figsize=(16, 4))
ax[0].violinplot([sepal_length[df['Species']==labels[0]], 
               sepal_length[df['Species']==labels[1]],
               sepal_length[df['Species']==labels[2]]])
ax[0].set_xticks([1, 2, 3])
ax[0].set_xticklabels([labels[0], labels[1], labels[2]])
ax[0].set_title(features[0])
ax[1].violinplot([sepal_width[df['Species']==labels[0]], 
               sepal_width[df['Species']==labels[1]],
               sepal_width[df['Species']==labels[2]]])
ax[1].set_xticks([1, 2, 3])
ax[1].set_xticklabels([labels[0], labels[1], labels[2]])
ax[1].set_title(features[1])
ax[2].violinplot([petal_length[df['Species']==labels[0]], 
               petal_length[df['Species']==labels[1]],
               petal_length[df['Species']==labels[2]]])
ax[2].set_xticks([1, 2, 3])
ax[2].set_xticklabels([labels[0], labels[1], labels[2]])
ax[2].set_title(features[2])
ax[3].violinplot([petal_width[df['Species']==labels[0]], 
               petal_width[df['Species']==labels[1]],
               petal_width[df['Species']==labels[2]]])
ax[3].set_xticks([1, 2, 3])
ax[3].set_xticklabels([labels[0], labels[1], labels[2]])
ax[3].set_title(features[3])
plt.show()
```

## 箱ひげ図とバイオリン図の違い

```python
fig,ax = plt.subplots(1,4, figsize=(12, 4))
ax[0].boxplot([sepal_length[df['Species']==labels[0]], 
               sepal_length[df['Species']==labels[1]],
               sepal_length[df['Species']==labels[2]]], labels=[labels[0], labels[1], labels[2]])
ax[0].violinplot([sepal_length[df['Species']==labels[0]], 
               sepal_length[df['Species']==labels[1]],
               sepal_length[df['Species']==labels[2]]])
ax[0].set_title(features[0])
ax[1].boxplot([sepal_width[df['Species']==labels[0]], 
               sepal_width[df['Species']==labels[1]],
               sepal_width[df['Species']==labels[2]]], labels=[labels[0], labels[1], labels[2]])
ax[1].violinplot([sepal_width[df['Species']==labels[0]], 
               sepal_width[df['Species']==labels[1]],
               sepal_width[df['Species']==labels[2]]])
ax[1].set_title(features[1])
ax[2].boxplot([petal_length[df['Species']==labels[0]], 
               petal_length[df['Species']==labels[1]],
               petal_length[df['Species']==labels[2]]], labels=[labels[0], labels[1], labels[2]])
ax[2].violinplot([petal_length[df['Species']==labels[0]], 
               petal_length[df['Species']==labels[1]],
               petal_length[df['Species']==labels[2]]])
ax[2].set_title(features[2])
ax[3].boxplot([petal_width[df['Species']==labels[0]], 
               petal_width[df['Species']==labels[1]],
               petal_width[df['Species']==labels[2]]], labels=[labels[0], labels[1], labels[2]])
ax[3].violinplot([petal_width[df['Species']==labels[0]], 
               petal_width[df['Species']==labels[1]],
               petal_width[df['Species']==labels[2]]])
ax[3].set_title(features[3])
plt.show()
```

```python
"""
箱ひげ図は中央値、四分位数、外れ値が一目でわかるところがよいが、分布の形は分からず、多峰性の分布では気を付けなければならない。
バイオリン図はデータの分布が一目でわかるところがよいが、正確な中央値の場所や、四分位数の場所は分からず、外れ値についても分かりづらい。
"""
```
