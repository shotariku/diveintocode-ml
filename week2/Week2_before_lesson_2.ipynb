# 【問題1】乱数の作成

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
two_dimensional_normal_distribution = np.random.multivariate_normal(mu, sigma, 500)
print(two_dimensional_normal_distribution)
```

# 【問題2】散布図による可視化

```python
plt.scatter(two_dimensional_normal_distribution[:,0], two_dimensional_normal_distribution[:,1])
```

# 【問題3】ヒストグラムによる可視化

```python
plt.xlim(-6, 3)
plt.hist(two_dimensional_normal_distribution[:,0], bins=50)
```

```python
plt.xlim(-6, 3)
plt.hist(two_dimensional_normal_distribution[:,1], bins=50)
```

# 【問題4】データの追加

```python
mu_small_y = [0, -3]
sigma = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
two_dimensional_normal_distribution_small_y = np.random.multivariate_normal(mu_small_y, sigma, 500)
print(two_dimensional_normal_distribution)
```

```python
plt.scatter(two_dimensional_normal_distribution[:,0], two_dimensional_normal_distribution[:,1], label=0)
plt.scatter(two_dimensional_normal_distribution_small_y[:,0], two_dimensional_normal_distribution_small_y[:,1], label=1)
plt.legend(loc='upper right', fontsize=10)
```
