# 【問題7】可視化

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
def decision_region(X, y, model, step=0.01, title='decision region', xlabel='sepal length (cm)', ylabel='petal length (cm)', 
                    target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2
    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), 
                                    np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred_str = model.predict(mesh).reshape(mesh_f0.shape)
    y_pred = np.where(y_pred_str=='versicolor', 1, 0)
    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
decision_region(va_x, va_y, knc_list[7])
```

# 【問題8】他の手法の学習

```python
def evaluation(tr_x, tr_y, va_x, va_y, model):
    """
    Accuracy, Precision, Recall, F値を計算する 

    Parameters
    ----------------
    tr_x : ndarray, shape(n_samples, 2)
          train_data 特徴量
    tr_y : ndarray, shape(n_samples,)
          train_data ラベル
    va_x : ndarray, shape(n_samples, 2)
          validationdata 特徴量
    va_y : ndarray, shape(n_samples,)
          validation_data ラベル     
    model : object
        学習したモデルのインスンタスを入れる
    """
    model.fit(tr_x, tr_y)
    y_pred = model.predict(va_x)
    accuracy = accuracy_score(va_y, y_pred)
    precision = precision_score(va_y, y_pred, pos_label='versicolor')
    recall = recall_score(va_y, y_pred, pos_label='versicolor')
    f1 = f1_score(va_y, y_pred, pos_label='versicolor')
    return accuracy, precision, recall, f1
```

```python
def evaluation_multi(tr_x, tr_y, va_x, va_y, model):
    """
    Accuracy, Precision, Recall, F値を計算する 

    Parameters
    ----------------
    tr_x : ndarray, shape(n_samples, 2)
          train_data 特徴量
    tr_y : ndarray, shape(n_samples,)
          train_data ラベル
    va_x : ndarray, shape(n_samples, 2)
          validationdata 特徴量
    va_y : ndarray, shape(n_samples,)
          validation_data ラベル     
    model : object
        学習したモデルのインスンタスを入れる
    """
    model.fit(tr_x, tr_y)
    y_pred = model.predict(va_x)
    accuracy = accuracy_score(va_y, y_pred)
    precision = precision_score(va_y, y_pred, average='macro')
    recall = recall_score(va_y, y_pred, average='macro')
    f1 = f1_score(va_y, y_pred, average='macro')
    return accuracy, precision, recall, f1
```

```python
def decision_region_multi(X, y, model, step=0.01, title='decision region', xlabel='sepal length (cm)', ylabel='petal length (cm)', 
                    target_names=['setosa', 'versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue', 'green']
    contourf_color = ['pink', 'springgreen',  'skyblue']
    n_class = 3
    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), 
                                    np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred_str = model.predict(mesh).reshape(mesh_f0.shape)
    y_pred_tmp = np.where(y_pred_str=='versicolor', 2, y_pred_str)
    y_pred_tmp_2 = np.where(y_pred_tmp=='virginica', 1, y_pred_tmp)
    y_pred = np.where(y_pred_tmp=='setosa', 0, y_pred_tmp_2)
    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

### 最も精度が高い手法は決定木とランダムフォレストであった。決定領域からSVMは過学習を起こしてしまったと考えられる。
